```python
import pandas as pd

dataset = pd.read_csv("https://gist.githubusercontent.com/adameubanks/35a6beea49e5b9ba62797e595a9626c0/raw/8974e055bdf3a9d7e6cacf1c1c30fcfd2ffd6de3/cancer.csv")
dataset
```

```python
x = dataset.drop(columns=['diagnosis(1=m, 0=b)'])
y = dataset['diagnosis(1=m, 0=b)']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
import tensorflow as tf

model = tf.keras.models.Sequential()
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(256, input_shape=(30,), activation="sigmoid"),
    tf.keras.layers.Dense(256, activation="sigmoid"),
    tf.keras.layers.Dense(1, activation="sigmoid")
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=1000)
```

```python
model.evaluate(x_test, y_test)
```
